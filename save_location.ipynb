```python
import googlemaps
import os.path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

from location import get_location_data
%matplotlib inline
```

```python
users_scifi = pd.read_csv(os.path.join("data","scifi.stackexchange.com", "Users.csv"))
```

```python
users_writers = pd.read_csv(os.path.join("data","writers.stackexchange.com", "Users.csv"))
users_movies = pd.read_csv(os.path.join("data","movies.stackexchange.com", "Users.csv"))
```

```python
def save_location(users, path):
    """Modifies data frame in place adding location data (lattitude, longitude, country and city)
    Saves data frame as a csv in current folder"""
    location_data = users["Location"].map(lambda location: get_location_data(location, True, True))
    location_data_clean = [x if (x is not None and len(x)==4) else tuple(np.nan for _ in range(4)) 
                           for x in location_data]
    users["lat"], users["lon"], users["country"], users["city"] = zip(*location_data_clean)
    users.to_csv(path)
```

```python
save_location(users_scifi, os.path.join("preprocessed_data", "users_scifi_location.csv"))
```

```python
save_location(users_writers, os.path.join("preprocessed_data", "users_writers_location.csv"))
save_location(users_movies, os.path.join("preprocessed_data", "users_movies_location.csv"))
```
